```python
#2.if ~ else ~조건문
a = int(input('숫자를 입력하세요:'))
if a < 100:
  print(f"{a}는 100 보다 작군요")
else:
  print(f"{a}는 100 보다 크군요")
```

```python
#3.중첩  if 조건문
a = int(input('숫자를 입력하세요:'))
if a < 100:
    if a<50:
        print(f"{a}는 50보다 작군요")
    else:
        print(f"{a}는 50보다 크고 100보다 작군요")
else:
    if a ==100:
        print(f"{a}는 100과 같군요")
    else:
        print(f"{a}는 100 보다 크군요")
```

```python
### 가위 바위 보 게임 난수 발생
import random
#random.randint(1,10)
a = random.choice(['가위', '바위','보'])
#a = input('숫자를 입력하세요:')
b = input('가위바위 보 입력:' )

if a == '가위':
    if b=='가위' : print(f"비겼습니다.")
    elif b=='바위' : print(f"이겼습니다.")
    else:  print(f"졌습니다.")
elif a == '바위':
    if b=='가위' : print(f"졌습니다.")
    elif b=='바위' : print(f"비겼습니다.")
    else:  print(f"이겼습니다.")
if a == '보':
    if b=='가위' : print(f"이겼습니다.")
    elif b=='바위' : print(f"졌습니다.")
    else:  print(f"비겼습니다.")
```

```python
a= int(input('a변의 길이를 입력하세요:'))
b= int(input('b변의 길이를 입력하세요:'))
c= int(input('c변의 길이를 입력하세요:'))

if a**2 + b**2 == c**2:
    print('직각 삼각형 입니다.')
else:
    print('직각 삼각형이 아닙니다.')
```

```python
#4. 다중 조건문(if~ elif)

#실습 학생 구분하기
age = int(input('나이를 입력하세요:'))
if age < 8:
    print('미취학 아동입니다.')
elif age < 14:
    print('초등 학생 입니다.')
elif age <17:
    print('중학생 입니다.')
elif age <20:
    print('고등학생 입니다.')
else:
    print('성인 입니다.')
```

```python
#5.조건부 표현식

score = int(input('점수를 입력하세요'))

if score >=80:
    messge = 'Success'
else:
    messge = 'Failure'

print(f'당신은 {messge}하셨습니다.')
```

```python
# - 온도가 30도 이상이면 에어컨을 켜고
# - 온도가 20도 미만이면 되면 내복을 입고
# - 온도가 10도 이하면 난방기를 켠다

a = int(input('온도를 입력하세요'))

if a >= 30:
    print('에어컨을 켜세요.')
elif a < 20 and a > 10:
    print('내복을 입으세요.')
elif  a <= 10:
    print('난방기를 켜세요.')
```

```python
menu = '''[커피주문 머신메뉴]
----------------------------
- 아메리카노 : 2500원
- 카페라떼   : 3000원
- 카페모카   : 3000원
원하는 커피 종류와 잔수를 입력하세요.'''

print(menu)
coffee1 = int(input('아메리카노 몇잔 : '))
coffee2 = int(input('카페라떼 몇잔 : '))
coffee3 = int(input('카페모카 몇잔 : '))
total = coffee1*2500 + coffee2*3000 + coffee3*3000
print(f'지불할 총 금액은 {total:,}원 입니다.')

money = int(input('돈을 입력해 주세요 :'))
change = total-money
if total > money:
    print(f'돈이 부족합니다.{change:,}을 넣어 주세요')
else:
    print(f'거스름 돈은 {abs(change):,}입니다. 커피 맛있게 드세요')

```
